Problem statement: Given the dataset of adv, we have to predict the sales based on TV ad exp.

```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import sklearn.metrics
```

```python
#Load data
data=pd.read_csv('/content/Advertising.csv')
print(data.head())
```

```python
data.info()
```

```python
#Relationship between TV and Sales
plt.scatter(data['TV'],data['sales'])
plt.xlabel('TV ad exp')
plt.ylabel('Sales')
plt.show()
```

```python
#Segrete Independent and dependent variables
x=data['TV'].values.reshape(-1,1)
y=data['sales'].values.reshape(-1,1)
print(type(x))
```

```python
print(x)
```

```python
#Train Algorithm
model=LinearRegression()
model.fit(x,y)
```

```python
#Make some predictions
y_pred=model.predict(x)
print(y_pred)
```

```python
plt.scatter(data['TV'],data['sales'])
plt.plot(data['TV'],y_pred,color='red')
plt.xlabel('TV ad exp')
plt.ylabel('Sales')
plt.show()
```

```python
#Coeficients
print(model.coef_)  #B1
print(model.intercept_) #B0
```

```python
#Evaluate Model
print("MSE:",sklearn.metrics.mean_squared_error(y,y_pred))
print("RMSE:",np.sqrt(sklearn.metrics.mean_squared_error(y,y_pred)))
```

```python
#prediction on unknown data
tv_add=int(input("Enter TV add exp:"))
print(f"Estimated Sales:{model.predict([[tv_add]])}K")
```
