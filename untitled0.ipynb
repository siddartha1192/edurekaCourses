```python
import pandas as pd
df = pd.read_csv("Bookings.csv")

df['days_to_departure'] = df['days_to_departure'].astype(float)

# Group by 'from_city' to compute average days booked in advance and number of departures
city_departure_stats = df.groupby('from_city').agg(
    avg_days_to_departure=('days_to_departure', 'mean'),
    total_departures=('from_city', 'count')
)
```

```python
city_departure_stats = city_departure_stats[city_departure_stats['total_departures'] >= 5]
top_3_advance_cities = city_departure_stats.sort_values(by='avg_days_to_departure', ascending=False).head(3)
```

```python

df = pd.read_csv("Bookings.csv")
df['booking_time'] = pd.to_datetime(df['booking_time'])
df['year_quarter'] = df['booking_time'].dt.to_period('Q')

quarterly_counts = df.groupby(['year_quarter', 'device_type_used']).size().unstack(fill_value=0)

quarterly_counts.index = quarterly_counts.index.to_timestamp()


quarterly_counts.plot(marker='o', figsize=(12, 6))
plt.title('Quarterly Bookings by Device Type')
plt.xlabel('Year - Quarter')
plt.ylabel('Number of Bookings')
plt.grid(True)
plt.legend(title='Device Type')
plt.tight_layout()
plt.show()
```
