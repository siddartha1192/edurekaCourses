Problem Statement: The dataset contains information about 3  brands namely US,Japan,Europe.other parameters are given about the cars.The Objective is to train a DT model which can predict the brand of the car

```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
#Read Data
car_df=pd.read_csv('/content/cars.csv')
car_df.head()
```

```python
print(car_df.info())
```

```python
#Check Null Values
car_df.isnull().sum()
```

```python
#Encoding Target variable
le=LabelEncoder()
car_df['brand']=le.fit_transform(car_df['brand'])
print(car_df)
```

```python
#Dropping 'year' feature
car_df.drop('year',axis=1,inplace=True)
print(car_df)
```

```python
#Segregating Independent and dependent variables
x=car_df.drop('brand',axis=1)
y=car_df['brand']
print(x)
print(y)
```

```python
#Scaling numeric Features
#sc=StandardScaler()
#x_scaled=sc.fit_transform(x)
#print(x_scaled)
```

```python
#Train and Test Split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#Train DT algorithm
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
#Plot Decision Tree made by algorithm
plt.figure(figsize=(15,10))
plot_tree(dt,filled=True,feature_names=x.columns)
plt.savefig('dtree.png',dpi=300)
plt.show()
```

```python
#Evaluate the model on Train
y_pred_train=dt.predict(x_train)
print(y_pred_train)
```

```python
print(classification_report(y_train,y_pred_train))
```

```python
#Evaluate the model on Test
y_pred_test=dt.predict(x_test)
print(y_pred_test)
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
dt1=DecisionTreeClassifier(criterion='entropy',max_depth=3)
dt1.fit(x_train,y_train)
```

```python
y_pred_train1=dt1.predict(x_train)
print(y_pred_train1)
print(classification_report(y_train,y_pred_train1))
```

```python
y_pred_test1=dt1.predict(x_test)
print(y_pred_test1)
print(classification_report(y_test,y_pred_test1))
```
